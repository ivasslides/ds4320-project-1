```python
import pandas as pd 
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
```

```python
data = pd.read_csv('final_df.csv')
```

```python
# One-hot encode cities/stations if you include it as a feature
df = pd.get_dummies(data, columns=['STATION'], drop_first=False)

df.head()
```

```python
X = df.drop(columns=['DATE', 'DOM_MW', 'Sunrise', 'Sunset', 'DailyWeather'])
y = df['DOM_MW']

# Split data into train and test sets
X_train, X_test, y_train, y_test, extreme_train, extreme_test = train_test_split(
    X, y, df['EWE'], test_size=0.2, random_state=42)
```

```python
rf = RandomForestRegressor(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)

# Predict on the test set
y_pred = rf.predict(X_test)
```

```python
# Overall performance
overall_rmse = mean_squared_error(y_test, y_pred) ** 0.5 
overall_rmse = round(overall_rmse, 4)
overall_r2 = r2_score(y_test, y_pred)
overall_r2 = round(overall_r2, 4)
print("Overall RMSE:", overall_rmse)
print("Overall R2:", overall_r2)

# Performance by weather type
normal_mask = extreme_test == 0
extreme_mask = extreme_test == 1

# Normal weather
n_rmse = mean_squared_error(y_test[normal_mask], y_pred[normal_mask]) ** 0.5
n_rmse = round(n_rmse, 4)
n_r2 = r2_score(y_test[normal_mask], y_pred[normal_mask])
n_r2 = round(n_r2, 4)

print("Normal Weather RMSE:", n_rmse)
print("Normal Weather R2:", n_r2)

# Extreme weather
e_rmse = mean_squared_error(y_test[extreme_mask], y_pred[extreme_mask]) ** 0.5
e_rmse = round(e_rmse, 4)
e_r2 = r2_score(y_test[extreme_mask], y_pred[extreme_mask])
e_r2 = round(e_r2, 4)
print("Extreme Weather RMSE:", e_rmse)
print("Extreme Weather R2:", e_r2)
```
