```python
import pandas as pd 
import numpy as np 
```

```python
c59 = pd.read_csv('cville59.csv')
c918 = pd.read_csv('cville918.csv')
d59 = pd.read_csv('dulles59.csv')
d918 = pd.read_csv('dulles918.csv')
nor59 = pd.read_csv('nor59.csv')
nor918 = pd.read_csv('nor918.csv')
ly59 = pd.read_csv('ly59.csv')
ly918 = pd.read_csv('ly918.csv')
```

```python
# drop columns don't need 
c59 = c59.drop(columns=['BackupDirection', 'BackupDistance', 'BackupDistanceUnit', 'BackupElements', 'BackupElevation', 'BackupElevationUnit', 'BackupEquipment', 'BackupLatitude', 'BackupLongitude', 'BackupName',
                        'AWND', 'CDSD', 'CLDD', 'DSNW', 'DYHF', 'DYTS', 'HDSD', 'HTDD','WindEquipmentChangeDate'])

c918 = c918.drop(columns=['BackupDirection', 'BackupDistance', 'BackupDistanceUnit', 'BackupElements', 'BackupElevation', 'BackupElevationUnit', 'BackupEquipment', 'BackupLatitude', 'BackupLongitude', 'BackupName',
                        'AWND', 'CDSD', 'CLDD', 'DSNW', 'DYHF', 'DYTS', 'HDSD', 'HTDD','WindEquipmentChangeDate'])

d59 = d59.drop(columns=['BackupDirection', 'BackupDistance', 'BackupDistanceUnit', 'BackupElements', 'BackupElevation', 'BackupElevationUnit', 'BackupEquipment', 'BackupLatitude', 'BackupLongitude', 'BackupName',
                        'AWND', 'CDSD', 'CLDD', 'DSNW', 'DYHF', 'DYTS', 'HDSD', 'HTDD','WindEquipmentChangeDate'])
d918 = d918.drop(columns=['BackupDirection', 'BackupDistance', 'BackupDistanceUnit', 'BackupElements', 'BackupElevation', 'BackupElevationUnit', 'BackupEquipment', 'BackupLatitude', 'BackupLongitude', 'BackupName',
                        'AWND', 'CDSD', 'CLDD', 'DSNW', 'DYHF', 'DYTS', 'HDSD', 'HTDD','WindEquipmentChangeDate'])

nor59 = nor59.drop(columns=['BackupDirection', 'BackupDistance', 'BackupDistanceUnit', 'BackupElements', 'BackupElevation', 'BackupElevationUnit', 'BackupEquipment', 'BackupLatitude', 'BackupLongitude', 'BackupName',
                        'AWND', 'CDSD', 'CLDD', 'DSNW', 'DYHF', 'DYTS', 'HDSD', 'HTDD','WindEquipmentChangeDate'])
nor918 = nor918.drop(columns=['BackupDirection', 'BackupDistance', 'BackupDistanceUnit', 'BackupElements', 'BackupElevation', 'BackupElevationUnit', 'BackupEquipment', 'BackupLatitude', 'BackupLongitude', 'BackupName',
                        'AWND', 'CDSD', 'CLDD', 'DSNW', 'DYHF', 'DYTS', 'HDSD', 'HTDD','WindEquipmentChangeDate'])

ly59 = ly59.drop(columns=['BackupDirection', 'BackupDistance', 'BackupDistanceUnit', 'BackupElements', 'BackupElevation', 'BackupElevationUnit', 'BackupEquipment', 'BackupLatitude', 'BackupLongitude', 'BackupName',
                        'AWND', 'CDSD', 'CLDD', 'DSNW', 'DYHF', 'DYTS', 'HDSD', 'HTDD','WindEquipmentChangeDate'])
ly918 = ly918.drop(columns=['BackupDirection', 'BackupDistance', 'BackupDistanceUnit', 'BackupElements', 'BackupElevation', 'BackupElevationUnit', 'BackupEquipment', 'BackupLatitude', 'BackupLongitude', 'BackupName',
                        'AWND', 'CDSD', 'CLDD', 'DSNW', 'DYHF', 'DYTS', 'HDSD', 'HTDD','WindEquipmentChangeDate'])
```

```python
cville = pd.concat([c59, c918], ignore_index=True).sort_values(by='DATE').reset_index(drop=True)
dulles = pd.concat([d59, d918], ignore_index=True).sort_values(by='DATE').reset_index(drop=True)
norfolk = pd.concat([nor59, nor918], ignore_index=True).sort_values(by='DATE').reset_index(drop=True)
lynchburg = pd.concat([ly59, ly918], ignore_index=True).sort_values(by='DATE').reset_index(drop=True)
```

```python
cville['DATE'] = pd.to_datetime(cville['DATE'])
cville['DATE'] = cville['DATE'].dt.floor('H')

dulles['DATE'] = pd.to_datetime(dulles['DATE'])
dulles['DATE'] = dulles['DATE'].dt.floor('H')

norfolk['DATE'] = pd.to_datetime(norfolk['DATE'])
norfolk['DATE'] = norfolk['DATE'].dt.floor('H')

lynchburg['DATE'] = pd.to_datetime(lynchburg['DATE'])
lynchburg['DATE'] = lynchburg['DATE'].dt.floor('H')


cville['DailyAverageDryBulbTemperature'] = (cville['DailyAverageDryBulbTemperature']
    .astype(str).str.replace('s', '', regex=False).astype(float))
dulles['DailyAverageDryBulbTemperature'] = (dulles['DailyAverageDryBulbTemperature']
    .astype(str).str.replace('s', '', regex=False).astype(float))
norfolk['DailyAverageDryBulbTemperature'] = (norfolk['DailyAverageDryBulbTemperature']
    .astype(str).str.replace('s', '', regex=False).astype(float))
lynchburg['DailyAverageDryBulbTemperature'] = (lynchburg['DailyAverageDryBulbTemperature']
    .astype(str).str.replace('s', '', regex=False).astype(float))
```

```python
cville.insert(0, "Entry_ID", range(1, len(cville) + 1))
dulles.insert(0, "Entry_ID", range(1, len(dulles) + 1))
norfolk.insert(0, "Entry_ID", range(1, len(norfolk) + 1))
lynchburg.insert(0, "Entry_ID", range(1, len(lynchburg) + 1))
```

```python
cville.to_csv('cville.csv', index=False)
dulles.to_csv('dulles.csv', index=False)
norfolk.to_csv('norfolk.csv', index=False)
lynchburg.to_csv('lynchburg.csv', index=False)
```
